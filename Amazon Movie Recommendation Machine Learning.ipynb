# Ques4

```python
dataset = df.melt(id_vars= df.columns[0],value_vars=df.columns[1:],
             var_name='Movie',value_name='rating')
print(dataset.shape)
dataset.head()
```

```python
from surprise.model_selection import train_test_split
from surprise import Dataset,Reader
```

```python
from surprise.model_selection import cross_validate
from surprise import SVD
```

```python
reader = Reader(rating_scale=(-1, 10))
data_f = Dataset.load_from_df(dataset.fillna(0), reader=reader)
```

Splitting of data into training and test data

```python
train_data,test_data = train_test_split(data_f, test_size=0.2)
```

# Ques5

```python
algo_used = SVD()
```

```python
algo_used.fit(train_data) 
```

# Ques6

```python
predict = algo_used.test(test_data)
```

```python
cross_validate(algo_used,data_f,measures=['RMSE','MAE'],cv=3,verbose=True)
```
